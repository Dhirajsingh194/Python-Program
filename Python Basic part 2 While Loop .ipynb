```python
#AND condition
a=10
b=20
if a==10 and b==20:
    print("a value is 10 and b is 20")
elif a==10 and b<20:
    print("Wrong")
else:
    print("This is else block")
```

```python
#OR condition
#AND condition
a=10
b=20
if a==10 or b==20:
    print("a value is 10 and b is 20")
elif a==10 and b<20:
    print("Wrong")
else:
    print("This is else block")
```
